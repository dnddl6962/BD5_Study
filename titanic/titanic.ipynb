```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.drop('Sex', axis=1, inplace=True)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df
```

```python
list1 = []
for i in range(df['Name'].count()):
    list1.append(df['Name'][i].split(', ')[1].split('. ')[0])
```

```python
df['Sex'] = list1
```

```python
df['Sex'].unique()
```

```python
sex_mapping = {
    'Mr': 'Male',
    'Master': 'Male',
    'Col': 'Male',
    'Rev': 'Male',
    'Dr': 'Male',  
    'Mrs': 'Female',
    'Miss': 'Female',
    'Ms': 'Female',
    'Dona': 'Female'
}
```

```python
df['Sex'] = df['Sex'].map(sex_mapping)
```

```python
df['Sex'].unique()
```

```python
df['Age'].isna().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Age'].isna().sum()
```

```python
df['Check'] = np.where((df['Sex']=='Male') & (df['Age']>=10) & (df['Age']<20), 'O', 'X')
```

```python
df.to_csv('result.csv', index=False)
```
